```python
'''
For this project, I will do some exploratory data analysis into this dataset
that looks into the job outcomes of students who graudated from college from 2010
and 2012. The data was released by American Community Survey, which conducts surveys
and aggregates the data. Each row in the dataset represents a different major in
college and contains information on gender diversity, employment rates, median salaries,
and more.

This time I will be working with Jupyter Notebook to show the visualization outputs
that I got from this project.
'''
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads=pd.read_csv('recent-grads.csv',index_col=0)
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
```

```python
raw_data_count=recent_grads.describe().iloc[0].max()
recent_grads=recent_grads.dropna()
recent_grads.describe()
cleaned_data_count=recent_grads.describe().iloc[0].max()
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample_size vs. Median')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs. Unemployment_rate')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs. Median')
```

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='ShareWomen vs. Median')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median')
```

```python
recent_grads.plot(x='Total', y='Median', kind='scatter', title='Major vs. Median')
```

```python
'''
From what we can see from all of these plots, students in the more popular
majors does not necessarily make more money than the other majors. In fact,
some of the best paying majors are actually less popular.

Also, from the Sharewomen vs Median plot, we can see that students that majored
in subjects that were majority female make less funny as we can see with the weak
down trend. We can see this happening in real life as well since STEM majors have
less women, but do tend to be higher paying than non-STEM majors.

By looking at the Full_time vs Median plot, we can also see that majors that have
less full-time students tend to be higher paying when students graduate. This is
probably because these majors with less full-time employees tend to be more rigorous
and thus require full-attention
'''
```

```python
'''
Let's see if we can get any insights if we shift from our method of 
scatter plots to histograms.
'''
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```
